```python
tarefas = []

id_tarefa = 1

while True:
  
    print("Selecione uma opção")
    print("1- Registrar uma nova Tarefa")
    print("2- Editar uma nova Tarefa")
    print("3- Marcar Tarefas como finalizadas")
    print("4- Listar Tarefas")
    print("5- Sair")
    opcao = input("Digite sua opção: ")

    if opcao == "1":
        
        while True:
            tarefa = input("Digite a tarefa: ")

            tarefas.append((id_tarefa, tarefa, "[ ]"))

            id_tarefa += 1

            continuar = input("Deseja continuar? (s/n): ")

            if continuar != "s":
                break

    elif opcao == "2":
     
        id_tarefa_editar = int(input("Digite o Índice da tarefa a editar: "))

        if 0 <= id_tarefa_editar < len(tarefas):
            
            tarefa_nova = input("Digite a nova tarefa: ")
            tarefas[id_tarefa_editar] = (id_tarefa_editar + 1, tarefa_nova, "[ ]")
        else:
            print("Índice da tarefa inválido.")

    elif opcao == "3":
       
        id_tarefa_marcar = int(input("Digite o Índice da tarefa a marcar como finalizada: "))

        if 0 <= id_tarefa_marcar < len(tarefas):
            # Marca a tarefa como finalizada
            tarefa_atual = list(tarefas[id_tarefa_marcar])
            tarefa_atual[2] = "[X]"
            tarefas[id_tarefa_marcar] = tuple(tarefa_atual)
            print(f"Tarefa {id_tarefa_marcar + 1} marcada como finalizada.")
        else:
            print("Índice da tarefa inválido.")

    elif opcao == "4":
        
        for i in tarefas:
            print(i)
    elif opcao == "5":
        
        print("Saindo...")
        break

    if opcao != "5":
        continue
```
