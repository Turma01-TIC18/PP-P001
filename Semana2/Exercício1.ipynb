```python
# Declara uma lista vazia para armazenar as tarefas
tarefas = []

# Declara uma variável para armazenar o ID da tarefa
id_tarefa = 1

# Loop para solicitar ao usuário que digite as tarefas
while True:
    # Solicita ao usuário que digite a opção
    print("Selecione uma opção")
    print("1- Registrar uma nova Tarefa")
    print("2- Editar uma nova Tarefa")
    print("3- Marcar Tarefas como finalizadas")
    print("4- Listar Tarefas")
    print("5- Sair")
    opcao = input("Digite sua opção: ")

    # Verifica a opção selecionada
    if opcao == "1":
        # Registra uma nova tarefa
        while True:
            tarefa = input("Digite a tarefa: ")

            # Adiciona a tarefa à lista
            tarefas.append((id_tarefa, tarefa, "[ ]"))

            # Incrementa o ID da tarefa
            id_tarefa += 1

            # Verifica se o usuário deseja continuar
            continuar = input("Deseja continuar? (s/n): ")

            # Se o usuário digitar "n", o loop é encerrado
            if continuar != "s":
                break

    elif opcao == "2":
        # Edita uma tarefa
        id_tarefa_editar = int(input("Digite o Índice da tarefa a editar: "))

        # Verifica se o ID da tarefa existe
        if 0 <= id_tarefa_editar < len(tarefas):
            # Edita a tarefa
            tarefa_nova = input("Digite a nova tarefa: ")
            tarefas[id_tarefa_editar] = (id_tarefa_editar + 1, tarefa_nova, "[ ]")
        else:
            print("Índice da tarefa inválido.")

    elif opcao == "3":
        # Marca tarefas como finalizadas
        id_tarefa_marcar = int(input("Digite o Índice da tarefa a marcar como finalizada: "))

        # Verifica se o Índice da tarefa existe
        if 0 <= id_tarefa_marcar < len(tarefas):
            # Marca a tarefa como finalizada
            tarefa_atual = list(tarefas[id_tarefa_marcar])
            tarefa_atual[2] = "[X]"
            tarefas[id_tarefa_marcar] = tuple(tarefa_atual)
            print(f"Tarefa {id_tarefa_marcar + 1} marcada como finalizada.")
        else:
            print("Índice da tarefa inválido.")

    elif opcao == "4":
        # Lista as tarefas
        #for id_tarefa, tarefa, finalizar in tarefas:
            #print(f"{id_tarefa}. {tarefa} {finalizar}")
        for i in tarefas:
            print(i)
    elif opcao == "5":
        # Sai do programa
        print("Saindo...")
        break

    # Se a opção não for "5", o loop continua
    if opcao != "5":
        continue
```
